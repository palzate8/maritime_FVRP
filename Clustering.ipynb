```python
!pip install ortools
!pip install geopy
!pip install scikit-learn
from ortools.constraint_solver import routing_enums_pb2
from ortools.constraint_solver import pywrapcp
import matplotlib.pyplot as plt
import math
import numpy as np
from geopy.distance import great_circle
from sklearn.base import BaseEstimator, ClusterMixin
from sklearn.cluster import KMeans
from sklearn.metrics import (
    calinski_harabasz_score,
    davies_bouldin_score,
    silhouette_score,
)
from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.spatial.distance import pdist


# Función para calcular la distancia geográfica entre dos puntos
def geo_distance(point1, point2):
    return great_circle(point1, point2).kilometers


# Función para calcular la matriz de distancias geográficas
def geo_distance_matrix(points):
    num_points = len(points)
    distances = np.zeros((num_points, num_points))
    for i in range(num_points):
        for j in range(i + 1, num_points):
            distances[i, j] = geo_distance(points[i], points[j])
            distances[j, i] = distances[i, j]
    return distances


# Matriz de dartos de los puertos (latitud, longitud)
points = np.array([
    [36.13326, -5.45051],  #Algeciras
    [23.73425, -99.11011],  #Altamira
    [51.1672, 4.2072],  #Antwerpen
    [8.9553, -79.5558],  #Balboa
    [39.29038, -76.61219],  #Baltimore
    [12.51106, -69.97422],  #Barcadera
    [41.38879, 2.15899],  #Barcelona
    [10.983, -74.95],  #Barranquilla
    [53.5396, 8.5789],  #Bremerhaven
    [3.96275, -77.37635],  #Buenaventura
    [-38.04354, -57.5458],  #Buenos Aires
    [35.10168, 129.03004],  #Busan
    [-12.05226, -77.13911],  #Callao
    [10.39972, -75.51444],  #Cartagena CONTECAR
    [10.40631, -75.52839],  #Cartagena SPRC
    [18.42423, -69.6345],  #Caucedo
    [32.77657, -79.93092],  #Charleston
    [22.4909, 113.8845],  #Chiwan
    [9.3589, -79.8997],  #Colon container terminal
    [9.3544, -79.9066],  #Cristobal
    [51.0306, 2.30202],  #Dunkirk
    [51.9551, 1.3110],  #Felixstowe
    [14.6, -61.08333],  #Fort de France
    [54.3659, 18.6591],  #Gdansk
    [44.4264, 8.91519],  #Genoa
    [-2.2785, -79.9079],  #Guayaquil
    [51.21989, 44.0346],  #Halifax
    [53.57532, 10.01534],  #Hamburg CTB
    [29.75757, -95.3512],  #Houston
    [17.99702, -76.79358],  #Kingston
    [10.59574, -66.95663],  #La guaira
    [49.49437, 0.108],  #Le Havre
    [35.88556, 14.5475],  #Leghorn
    [51.5072, -0.1275],  #London Gateway
    [34.05223, -118.24368],  #Los angeles
    [0.03889, -51.06639],  #Macapa
    [-3.10194, -60.025],  #Manaus
    [19.11695, -104.34214],  #Manzanillo (Mx)
    [9.3597, -79.8828],  #Manzanillo International Terminal (MIT)
    [22.9919, -82.7667],  #Mariel
    [35.84194, 14.54306],  #Marsaxlok
    [43.29695, 5.38107],  #Marseilles Fos
    [-37.8242, 144.9679],  #Melbourne
    [25.77417, -80.17111],  #Miami
    [10.0042, -83.0767],  #Moin APM Terminals
    [-34.90111, -56.16453],  #Montevideo
    [47.3266, -2.14998],  #Montoir de Bretagne
    [-39.48333, 176.91667],  #Napier
    [-26.8994, -48.6553],  #Navegantes
    [29.95465, -90.07507],  #New Orleans
    [40.71427, -74.00597],  #New york
    [29.87819, 121.62039],  #Ningbo
    [36.84681, -76.28522],  #Norfolk
    [37.7699, -122.226],  #Oakland
    [-5.0817, -81.177],  #Paita
    [-25.50469, -48.51082],  #Paranagua
    [39.95233, -75.16379],  #Philadelphia
    [16.23824, -61.53211],  #Pointe a Pitre
    [-45.8159, 170.62135],  #Port chalmers
    [25.8576, -81.3866],  #Port Everglades
    [10.67107, -61.52121],  #Port of Spain
    [-26.17904, -48.60622],  #Poto Itapoa
    [15.72778, -88.59444],  #Puerto Barrios
    [10.4667, -68.0167],  #Puerto Cabello
    [15.82562, -87.92968],  #Puerto Cortes
    [-2.71748, -80.30042],  #Puerto de Posorja
    [36.0247, 120.357],  #Qingdao
    [-22.90642, -43.18223],  #Rio de Janeiro
    [-30.0277, -51.2287],  #Rio Grande
    [18.417, -70.019],  #Rio Haina
    [51.92423, 4.50125],  #Rotterdam
    [-12.97111, -38.51083],  #Salvador
    [-33.5933, -71.6217],  #San Antonio (Chile)
    [18.46633, -66.10572],  #San Juan (Puerto Rico)
    [11.24079, -74.19904],  #Santa Marta
    [-23.96083, -46.33361],  #Santos
    [32.08354, 81.09983],  #Savannah
    [31.22222, 121.45806],  #Shanghai
    [1.25937, 103.75445],  #Singapure
    [50.897, -1.40416],  #Southampton
    [-8.39438, -34.97405],  #Suape
    [-33.86785, 151.20732],  #Sydney
    [-37.68611, 176.16667],  #Tauranga
    [8.9369, -79.5755],  #Terminal internacional PSA Panamá
    [39.46975, -0.37739],  #Valencia
    [-33.03932, -71.62625],  #Valparaiso
    [19.4795, -99.08862],  #Veracruz
    [41.3581, -8.74756],  #Vila do Conde
    [12.1084, -68.93354],  #Willemstad
    [51.3157, 3.2070]  #Zeebrugge
])

# Calcular la matriz de distancias geográficas
distances = geo_distance_matrix(points)

#DEFINICIÓN DE MÉTODO PARA CALCULAR LA DISTANCIA DE HAVERSINE ENTRE DOS PUNTOS


# Implementación de la distancia de Haversine
def haversine_distance(coord1, coord2):
    R = 6371.0  # Radio de la Tierra en kilómetros
    lat1, lon1 = np.deg2rad(coord1)
    lat2, lon2 = np.deg2rad(coord2)

    dlat = lat2 - lat1
    dlon = lon2 - lon1

    a = np.sin(dlat / 2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))
    return R * c


# Implementación de K-means usando la distancia de Haversine
class HaversineKMeans(BaseEstimator, ClusterMixin):

    def __init__(self,
                 n_clusters=3,
                 max_iter=300,
                 tol=1e-4,
                 random_state=None):
        self.n_clusters = n_clusters
        self.max_iter = max_iter
        self.tol = tol
        self.random_state = random_state

    def fit(self, X):
        np.random.seed(self.random_state)
        self.cluster_centers_ = X[np.random.choice(X.shape[0],
                                                   self.n_clusters,
                                                   replace=False)]

        for _ in range(self.max_iter):
            distances = self.haversine_distance(X, self.cluster_centers_)
            self.labels_ = np.argmin(distances, axis=1)

            new_centers = np.array([
                X[self.labels_ == i].mean(axis=0) if np.any(
                    self.labels_ == i) else self.cluster_centers_[i]
                for i in range(self.n_clusters)
            ])

            if np.all(np.abs(new_centers - self.cluster_centers_) < self.tol):
                break

            self.cluster_centers_ = new_centers

        return self

    def haversine_distance(self, X, Y):

        def haversine(coord1, coord2):
            R = 6371.0  # Radio de la Tierra en kilómetros
            lat1, lon1 = np.deg2rad(coord1)
            lat2, lon2 = np.deg2rad(coord2)

            dlat = lat2 - lat1
            dlon = lon2 - lon1

            a = np.sin(dlat / 2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(
                dlon / 2)**2
            c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))
            return R * c

        return np.array([[haversine(x, y) for y in Y] for x in X])


# Aplicar el algoritmo HaversineKMeans con k de prueba

# Visualizar los resultados del clustering con k de prueba

hkm = HaversineKMeans(n_clusters=5, random_state=42)
hkm.fit(points)

# Resultados
print("Centroides de clusters:")
print(hkm.cluster_centers_)
print("Etiquetas de los puntos:")
print(hkm.labels_)


def plot_clusters(coords, labels, centers):
    plt.figure(figsize=(10, 7))
    plt.scatter(coords[:, 1],
                coords[:, 0],
                c=labels,
                cmap='viridis',
                s=100,
                edgecolors='k')
    plt.scatter(centers[:, 1],
                centers[:, 0],
                c='red',
                s=200,
                marker='x',
                label='Centroides')
    plt.xlabel('Longitud')
    plt.ylabel('Latitud')
    plt.title('Clusters geográficos')
    plt.legend()
    plt.grid(True)
    plt.show()


plot_clusters(points, hkm.labels_, hkm.cluster_centers_)


#CODIGO PARA APLICAR COEFICIENTES DE MEDICION DE CALIDAD DE CLÚSTER CON KMEANS GEOGRAFICO

# Calcular para diferentes valores de K
#inertia = [] no se aplica ya que el método del codo es incompatible con la distancia de Haversine

silhouette_scores = []
davies_bouldin_scores = []
calinski_harabasz_scores = []

K = range(8, 15)  # El rango se ajusta según las necesidades

#Teniendo en cuenta las características del problema y el tipo de datos se determina que, ya que los datos corresponden a puertos distribuidos en el planeta
#no tiene sentido que se realice una clusterización con menos de 6 clusters ya que, eso podría generar, en teoría, un clúster por continente y habría
#"sobrepoblación" de clústers en América

for k in K:

    kmeans = HaversineKMeans(n_clusters=k, random_state=42)
    # Random_state 42 semilla para controlar aleatoriedad 42 referencia cultural a "The Hitchhiker's Guide to the Galaxy" buscando respuesta a la pregunta        # fundamental sobre la vida, el universo y Todo lo demás.
    kmeans.fit(points)
    clusters = kmeans.fit_predict(points)
    silhouette_avg = silhouette_score(points, clusters)
    silhouette_scores.append(silhouette_avg)
    davies_bouldin = davies_bouldin_score(points, clusters)
    calinski_harabasz = calinski_harabasz_score(points, clusters)
    davies_bouldin_scores.append(davies_bouldin)
    calinski_harabasz_scores.append(calinski_harabasz)

#Gráfica de la curva de coeficiente de Calinski para número optimo de Clústers
plt.figure(figsize=(8, 5))
plt.plot(K, calinski_harabasz_scores, 'bx-')
plt.xlabel('Número de clusters K')
plt.ylabel('Índice de Calinski-Harabasz con Kmeans Geográfico')
plt.title(
    'Índice de Calinski-Harabasz para determinar el número óptimo de clusters')
plt.show()

#Gráfica de la curva de coeficiente de Davies-Bouldin para número optimo de Clústers
plt.figure(figsize=(8, 5))
plt.plot(K, davies_bouldin_scores, 'bx-')
plt.xlabel('Número de clusters K')
plt.ylabel('Índice de Davies-Bouldin con Kmeans Geográfico')
plt.title(
    'Índice de Davies-Bouldin para determinar el número óptimo de clusters')
plt.show()

#Gráfica de la curva de coeficiente de Silueta para número optimo de Clústers
plt.figure(figsize=(8, 5))
plt.plot(K, silhouette_scores, 'bx-')
plt.xlabel('Número de clusters K')
plt.ylabel('Coeficiente de Silueta con Kmeans Geográfico')
plt.title(
    'Coeficiente de Silueta para determinar el número óptimo de clusters')
plt.show()

print("Coeficientes de Calinski:")
print(calinski_harabasz_scores)
print("Coeficientes de Davies-Bouldin:")
print(davies_bouldin_scores)
print("Coeficientes de Silueta:")
print(silhouette_scores)

#Deteremina los valores óptimos según cada uno de las métricas (Calinski, DB, Silueta)

k_Cal = calinski_harabasz_scores.index(max (calinski_harabasz_scores)) + min(K)
print("Máximo Coeficientes de Calinski:")
print(k_Cal)

k_DB = davies_bouldin_scores.index(min (davies_bouldin_scores)) + min(K)
print("Mínimo Coeficientes de DB:")
print(k_DB)

k_Sil = silhouette_scores.index(max (silhouette_scores)) + min(K)
print("Máximo Coeficientes de Silueta:")
print(k_Sil)

k_optimo = math.floor((k_Cal+k_DB+k_Sil)/3)
print("K óptimo:")
print(k_optimo)

# Aplicación de KMeans Geográfico con el óptimo de clústers (promedio de los valores entregados por las métricas calculadas (Calinski, DB, Silueta))

kmeans = HaversineKMeans(n_clusters=k_optimo, random_state=42)
labels = kmeans.fit_predict(points)

print("Centroides de clusters:")
print(kmeans.cluster_centers_)
print("Etiquetas de los puntos:")
print(kmeans.labels_)

# Visualizar los resultados del clustering

plt.scatter(points[:, 1], points[:, 0], c=labels, cmap='viridis', s=100)
plt.xlabel('Longitud')
plt.ylabel('Latitud')
plt.title('Clustering KMeans Geográfico')
plt.show()

print(labels)

#Separación de los clusters según la asignación hecha por kmeans geográfico (cada cluster genera una matriz)

# Dada la matriz de datos "points" y las etiquetas de los clusters "labels"

# Encontrar las etiquetas únicas
etiquetas_unicas = np.unique(labels)

# Crear un diccionario para almacenar las submatrices
matrices_clusters = {etiqueta: [] for etiqueta in etiquetas_unicas}

# Llenar el diccionario con submatrices
for i, etiqueta in enumerate(labels):
    matrices_clusters[etiqueta].append(points[i])

# Convertir las listas en matrices numpy
for etiqueta in matrices_clusters:
    matrices_clusters[etiqueta] = np.array(matrices_clusters[etiqueta])

# Imprimir las submatrices
for etiqueta, submatriz in matrices_clusters.items():
    print(f"Submatriz para Clúster {etiqueta}:\n{submatriz}\n")


#GRAFICAR CADA UNA DE LOS CLUSTERS (SUMATRICES)

for etiqueta, submatriz in matrices_clusters.items():

    plt.scatter(submatriz[:, 1], submatriz[:, 0], c='blue', s=100)
    plt.xlabel('Longitud')
    plt.ylabel('Latitud')
    plt.title(f'Gráfico de puntos para Clúster {etiqueta}')
    plt.show()


# Función para calcular la matriz de distancias geográficas

def calcular_matriz_distancias(coordenadas):
    num_puntos = coordenadas.shape[0]
    matriz_distancias = np.zeros((num_puntos, num_puntos))

    for i in range(num_puntos):
        for j in range(num_puntos):
            if i != j:
                # Calcular la distancia entre los puntos i y j
                distancia = great_circle(coordenadas[i], coordenadas[j]).kilometers
                matriz_distancias[i, j] = distancia

    return matriz_distancias

# Función para procesar múltiples submatrices
def procesar_submatrices(submatrices):
    matrices_distancias = []

    for idx, submatriz in enumerate(submatrices):
        print(f"Calculando matriz de distancias para Clúster {idx}")
        matriz_distancias = calcular_matriz_distancias(submatriz)
        matrices_distancias.append(matriz_distancias)

    return matrices_distancias

# Obtener las submatrices como una lista
lista_submatrices = list(matrices_clusters.values())

# Procesar las submatrices
matrices_distancias = procesar_submatrices(lista_submatrices)

# Imprimir las matrices de distancias
for idx, matriz in enumerate(matrices_distancias):
    print(f"\nMatriz de distancias para Clúster {idx}:")
    print(matriz)


#####IMPLEMENTACIÓN DEL VRP

# Función para resolver el VRP
def resolver_vrp(matriz_distancias, num_vehiculos, deposito):
    # Crear el gestor de rutas
    manager = pywrapcp.RoutingIndexManager(len(matriz_distancias),
                                             num_vehiculos, deposito)

    # Crear el modelo de ruteo
    routing = pywrapcp.RoutingModel(manager)

    # Crear la función de distancia
    def distancia_callback(from_index, to_index):
        return int(matriz_distancias[manager.IndexToNode(from_index)][manager.IndexToNode(to_index)])

    transit_callback_index = routing.RegisterTransitCallback(distancia_callback)
    routing.SetArcCostEvaluatorOfAllVehicles(transit_callback_index)

    # Configurar parámetros de búsqueda
    search_parameters = pywrapcp.DefaultRoutingSearchParameters()
    search_parameters.first_solution_strategy = (
        routing_enums_pb2.FirstSolutionStrategy.PATH_CHEAPEST_ARC)

    # Resolver el problema
    solución = routing.SolveWithParameters(search_parameters)

    if solución:
        rutas = []
        for vehiculo in range(num_vehiculos):
            index = routing.Start(vehiculo)
            ruta = []
            while not routing.IsEnd(index):
                ruta.append(manager.IndexToNode(index))
                index = solución.Value(routing.NextVar(index))
            rutas.append(ruta)
        return rutas
    else:
        print('No se encontró solución.')
        return None


# Función para graficar las rutas
def graficar_rutas(rutas, coordenadas):
    plt.figure(figsize=(10, 8))

    # Graficar todos los puntos
    plt.scatter(coordenadas[:, 1], coordenadas[:, 0], color='black', label='Puntos')

    # Graficar las rutas
    colores = ['blue', 'green', 'red', 'purple', 'orange']
    for idx, ruta in enumerate(rutas):
        ruta_coords = coordenadas[ruta]
        plt.plot(ruta_coords[:, 1], ruta_coords[:, 0], marker='o', color=colores[idx % len(colores)], label=f'Ruta {idx + 1}')

    plt.xlabel('Longitud')
    plt.ylabel('Latitud')
    plt.title('Rutas de VRP')
    plt.legend()
    plt.grid(True)
    plt.show()

# Resolver el VRP para cada submatriz y graficar
num_vehiculos = 1
depot = 0

for etiqueta, submatriz in matrices_clusters.items():
    print(f"Resolviendo VRP para Clúster {etiqueta}")
    matriz_distancias = calcular_matriz_distancias(submatriz)
    rutas = resolver_vrp(matriz_distancias, num_vehiculos, depot)
    if rutas:
        print(f"Rutas para Clúster {etiqueta}: {rutas}")
        graficar_rutas(rutas, submatriz)

# Calcular la distancia euclidiana entre los puntos
distances = pdist(points, metric='euclidean')

# Realizar la clusterización jerárquica usando el método WARD (Criterio de varianza mínima: minimiza la varianza total dentro de los clúster)
Z = linkage(distances, method='ward')

# Visualizar el dendrograma
plt.figure(figsize=(10, 7))
dendrogram(Z, labels=range(1, len(points) + 1), leaf_rotation=90)
plt.title('Clusterización jerárquica de los puertos')
plt.xlabel('Puertos')
plt.ylabel('Distancia')
plt.show()
```
